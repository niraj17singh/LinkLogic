```python
from qwikidata.linked_data_interface import get_entity_dict_from_api
```

```python
def get_name_from_entity_dict(entity_dict):
    success = False
    if 'en' in entity_dict['labels'].keys():
        out = entity_dict['labels']['en']['value']
        success = True
    else:
        #print(f'Warning: using id instead of name.')
        out = entity_dict['id']
    return out, success
```

```python
import pandas as pd

rels = pd.read_csv('/Users/rhodos/Desktop/Projects/akbc2021/data/ogbl_wikikg2/mapping/reltype2relid.csv')
```

```python
rel_ids = list(rels['rel id'])
```

```python
rel_names = []
for rel in rel_ids:
    
    try:
        entity_dict = get_entity_dict_from_api(rel)
    except:
        rel_names.append(rel)
        print(f'warning: appending rel id instead of name for {rel}')
        continue
        
    name, success = get_name_from_entity_dict(entity_dict)
    if not success:
        print('warning: did not get name successfully')
    else:
        print(name)
    rel_names.append(name)
```

```python
rel_info = pd.DataFrame({'name': rel_names, 'id': rel_ids})
```

```python
rel_info.to_csv('rel_info.tsv', sep='\t', index=None)
```
