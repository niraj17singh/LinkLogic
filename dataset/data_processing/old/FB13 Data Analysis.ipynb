```python
data = []
relations_dict = {}
profession = []
c = 0
with open("train.txt", "r") as file:
    for f in file:
        f = f.strip()
        e1, r, e2 = f.split("\t")
        if r == "spouse":
            print(e1, r, e2)
            c+=1
            profession.append(e2)
        if r not in relations_dict:
            relations_dict[r] = 1
        else:
            relations_dict[r] += 1
        data.append([e1, r, e2])
relations_dict = {k: v for k, v in sorted(relations_dict.items(), key=lambda item: item[1], reverse=True)}    

        
for r in relations_dict:
    relations_dict[r] = relations_dict[r]
plt.bar(*zip(*relations_dict.items()));
plt.xticks(rotation = 90);
plt.xlabel("Relationship type")
plt.ylabel("Frequency distribution (Normalized)")
plt.title("FB13 - Test")
```

```python
data = []
relations_dict = {}
with open("valid.txt", "r") as file:
    for f in file:
        f = f.strip()
        e1, r, e2, l = f.split("\t")
        if r not in relations_dict:
            relations_dict[r] = 1
        else:
            relations_dict[r] += 1
        data.append([e1, r, e2, l])
relations_dict = {k: v for k, v in sorted(relations_dict.items(), key=lambda item: item[1], reverse=True)}    
for r in relations_dict:
    relations_dict[r] = relations_dict[r]/len(data)
plt.bar(*zip(*relations_dict.items()));
plt.xticks(rotation = 90);
```
