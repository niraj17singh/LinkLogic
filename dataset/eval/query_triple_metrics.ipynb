```python
import pandas as pd
import json
```

```python
triple_metrics = pd.read_csv('metrics_all_parents_triples.tsv', sep='\t')
```

```python
# load benchmark
with open('../../data/commonsense_benchmark/v4/commonsense_benchmark_all.json', 'r') as f:
    benchmark = json.load(f)
```

```python
rows = []
for bmk in benchmark:
    rows.append({
        'query triple': '--'.join(bmk['query_triple']),
        'num_siblings': len(bmk['entity_names']['siblings'])
    })
num_siblings = pd.DataFrame(rows)
```

```python
out = triple_metrics.merge(num_siblings, on='query triple', how='left')
```
